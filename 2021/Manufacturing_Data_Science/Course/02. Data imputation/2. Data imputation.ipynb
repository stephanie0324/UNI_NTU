## 3.插值法
### 以上面和下面一筆資料取一個中間值填補遺漏值，較適合用在連續性/ 時間序列性的資料

### 先判斷哪些column中有NA值

```python
na_col = []
for i in df.columns:
    if df[i].isnull().values.any():
        na_col.append(i)
na_col
```

```python
df_fillin = df.copy()
for i in na_col:
    df_fillin[i] = df[i].interpolate(method='linear')
```

```python
df_fillin.head(10)
```

## 4. 用K-Nearest Neighbours填補遺漏值

### 利用每個資料點算出個別的距離，並且得到最近的幾組資料(鄰居)依照這幾筆資料進行遺漏值填補

```python
X = df.drop(["TypeOfSteel_A300","TypeOfSteel_A400","Fault_class"],axis=1)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)
df_fillk = imputer.fit_transform(X)
```

```python
df_fillk = pd.DataFrame(df_fillk,columns=X.columns)
```

```python
df_fillk.head(10)
```

## 5. 用MICE填補遺漏值

### 用liner regression的方法把有遺漏值的欄位當作Y其他當作X下去建模，每項都這樣以此類推最後訓練好模型用預測的方法填補遺漏值

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=100, random_state=0)
imp.fit(X)
```

```python
df_fmice = imp.transform(X)
df_fmice = pd.DataFrame(df_fmice, columns=X.columns)
```

```python
df_fmice.head(10)
```

## 6. 總結

### 以上介紹了許多填補遺漏值的方法，在 sklearn.impute 這個套件裡也有更多的方法可以用，不過沒有哪一個方法是最好的，應該是要視不同的資料及不同的問題來看它適合處理遺漏值的方法，像是當今天有一筆資料有 80% 都是遺漏值，那可能刪除這筆資料比較適合，當另一種狀況，資料有很少筆，且取得資料相當困難，則填補遺漏值就相當重要。
